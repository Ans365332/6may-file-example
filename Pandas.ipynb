#Stacking-->
It is A way of dreshaping the dataframe by turning the column into row
and change into series with index 0

```python
df1=pd.DataFrame({
    'A':['Apple','Banana','Mango'],
    'B':[1,2,3],
    'C':[5,6,7]
})
df1
```

0,1,2 are index and A,B,C are labels

```python
df=df1.stack()
df
```

```python
df.unstack()
```
