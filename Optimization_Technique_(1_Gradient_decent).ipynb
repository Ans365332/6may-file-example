# Optimization technique -->
It is used for minimize the error of prediction and actual value.



```python
# (1). Gradient Descent ::::

# Loss function= y_act - y_pred  ( for single value we have loss function and for entire dataset we have cost function)
```

```python
# We Konow the Equation of simple line :::
# y=mx+b .....(1)   where (x,y) -->coordinates,m --> slope,b -->intercept


# y_pred = mx+b
# y_act(y_test) = mx+b

# So we will calculate Loss function w.r.t Mean squared error :::
# L = 1/n * (y_act - y_pred)**2    -----(2)

# Now,if we have single feature then (n=1)
# L = (y_act - y_pred)**2  ------(3)

# L = (y-(mx+b))**2   -----(4)
# L = (y-mx-b)**2  ------(5)

# So,we will differentiate Loss w.r.t Intercept (b)

# dL/db = 2(y-mx-b).[dy/db - dmx/db - db/db]
# dL/db = 2(y-mx-b).[0 - 0 - 1]
# dL/db = -2(y-mx-b)   ------(6)  (location of intercept)

# Now,we will differentiate Loss w.r.t slope (m)

# dL/dm = 2(y-mx-b).[dy/dm - dmx/dm - db/dm]
# dL/dm = 2(y-mx-b).[0 - x - 0]
# dL/dm = -2x(y-mx-b)  -------(7)  (location of slope)


# Now,
# b_new = b_old - learning_rate * dL/db  (updated intercept)
# m_new = m_old - learning_rate * dL/dm  (updated slope)


# if lr >>>>> 0 ----> Gradient Exploading
# if lr <<<<< 0 ----> Vanishing Gradient
# Appropriate value of lr = 0.001


# So,our final Equation of new line will be ::::

# from Equa(8,9)
# y = m_new * X + b_new  -----(10)

# this is the best optimized solution and it reuce the error or minimize the loss of line
```

```python
epochs = 100
m = 78.35
lr = 0.1

def gradient_descent(X, y, m, b, lr, epochs):
    b_history = [b]
    y_pred_history = []

    for i in range(epochs):
        y_pred = m * X.ravel() + b
        loss_slope = -2 * np.sum(y - y_pred)
        step_size = loss_slope * lr
        b = b - step_size
        b_history.append(b)
        y_pred_history.append(m * X.ravel() + b)

    return b, b_history, y_pred_history
```

```python
b_initial = 200 # Starting with a different initial 'b'
b_final, b_hist, y_pred_hist = gradient_descent(X, y, m, b_initial, lr, epochs)

print(f"Initial intercept (b): {b_initial}")
print(f"Final intercept (b) after {epochs} epochs: {b_final}")
print(f"OLS intercept: {reg.intercept_}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Actual Data')
plt.plot(X, reg.predict(X), color='black', label='OLS Regression Line')

# Plotting the progression of the gradient descent
for i in range(0, epochs, 10): # Plot every 10th iteration
    plt.plot(X, y_pred_hist[i], color='orange', linestyle='--', alpha=0.5, label=f'GD Iteration {i+1}')

plt.plot(X, y_pred_hist[-1], color='red', linewidth=2, label=f'GD Final Line (Epoch {epochs})')

plt.title('Gradient Descent Convergence for Intercept')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(epochs + 1), b_hist)
plt.axhline(y=reg.intercept_, color='black', linestyle='--', label='OLS Intercept')
plt.title('Intercept (b) Value over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Intercept (b)')
plt.legend()
plt.grid(True)
plt.show()
```
