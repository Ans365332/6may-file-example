```python
# without column transformer
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('/content/insurance - insurance (1).csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['region'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
x_train.shape
```

```python
oe=OrdinalEncoder(categories=[['male','female'],['no','yes'],['southwest','southeast','northwest','northeast']])

x_train_sex_smoker_region=oe.fit_transform(x_train[['sex','smoker','region']])

# for test data
x_test_sex_smoker_region=oe.fit_transform(x_test[['sex','smoker','region']])

x_train_sex_smoker_region.shape
```

```python
# now extract column

x_train_age = x_train[['age']].values
x_test_age = x_test[['age']].values
x_train_age.shape
```

```python
x_train_bmi = x_train[['bmi']].values
x_test_bmi = x_test[['bmi']].values
x_train_bmi.shape
```

```python
x_train_children = x_train[['children']].values
x_test_children = x_test[['children']].values
x_train_children.shape
```

```python
x_train_transformed=np.concatenate((x_train_age,x_train_bmi,x_train_children,x_train_sex_smoker_region),axis=1)
```

```python
x_test_transformed=np.concatenate((x_test_age,x_test_bmi,x_test_children,x_test_sex_smoker_region),axis=1)
```

```python
x_train_transformed.shape
```

```python
# with column transformer
```

```python
from sklearn.compose import ColumnTransformer

transformer=ColumnTransformer(transformers=[
    ('tnf1',OrdinalEncoder(categories=[['male','female']]),['sex']),
    ('tnf2',OrdinalEncoder(categories=[['no','yes']]),['smoker']),
    ('tnf3',OrdinalEncoder(categories=[['southwest','southeast','northwest','northeast']]),['region'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train).shape
```

```python
transformer.fit_transform(x_test).shape
```

```python
df=pd.read_csv('/content/credit_scoring - credit_scoring.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Employment Status'].value_counts()
```

```python
x=df.drop(columns=['Credit Utilization Ratio'])
y=df['Credit Utilization Ratio']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
# ordinal encoding -->

oe=OrdinalEncoder(categories=[['Male','Female'],['Married','Single','Divorced'],['Master','High School','PhD','Bachelor']
                              ,['Self-Employed','Employed','Unemployed'],['Auto Loan','Home Loan','Personal Loan']])
x_train_s_m_e_ma_l=oe.fit_transform(x_train[['Gender','Marital Status','Education Level','Employment Status','Type of Loan']])

x_test_s_m_e_ma_l=oe.fit_transform(x_test[['Gender','Marital Status','Education Level','Employment Status','Type of Loan']])

x_train_s_m_e_ma_l.shape
```

```python
# extract

x_train_Age = x_train[['Age']].values
x_test_Age = x_test[['Age']].values
x_train_Age.shape
```

```python
x_train_Payment_History = x_train[['Payment History']].values
x_test_Payment_History = x_test[['Payment History']].values
x_train_Payment_History.shape
```

```python
x_train_Number_of_Credit_Accounts = x_train[['Number of Credit Accounts']].values
x_test_Number_of_Credit_Accounts = x_test[['Number of Credit Accounts']].values
x_train_Number_of_Credit_Accounts.shape
```

```python
x_train_Loan_Amount = x_train[['Loan Amount']].values
x_test_Loan_Amount = x_test[['Loan Amount']].values
x_train_Loan_Amount.shape
```

```python
x_train_Interest_Rate= x_train[['Interest Rate']].values
x_test_Interest_Rate = x_test[['Interest Rate']].values
x_train_Interest_Rate.shape
```

```python
x_train_Loan_Term = x_train[['Loan Term']].values
x_test_Loan_Term = x_test[['Loan Term']].values
x_train_Loan_Term.shape
```

```python
x_train_transformed = np.concatenate((x_train_Age,x_train_Payment_History,x_train_Number_of_Credit_Accounts,
                                      x_train_Loan_Amount,x_train_Interest_Rate,x_train_Loan_Term,x_train_s_m_e_ma_l),axis=1)
```

```python
x_test_transformed = np.concatenate((x_test_Age,x_test_Payment_History,x_test_Number_of_Credit_Accounts,
                                      x_test_Loan_Amount,x_test_Interest_Rate,x_test_Loan_Term,x_test_s_m_e_ma_l),axis=1)
```

```python
x_train_transformed.shape
```

```python
# with column transformer
```

```python
transformed=ColumnTransformer(transformers=[
    ('tnf1',OrdinalEncoder(categories=[['Male','Female']]),['Gender']),
    ('tnf2',OrdinalEncoder(categories=[['Married','Single','Divorced']]),['Marital Status']),
    ('tnf3',OrdinalEncoder(categories=[['Master','High School','PhD','Bachelor']]),['Education Level']),
    ('tnf4',OrdinalEncoder(categories=[['Self-Employed','Employed','Unemployed']]),['Employment Status']),
    ('tnf5',OrdinalEncoder(categories=[['Auto Loan','Home Loan','Personal Loan']]),['Type of Loan'])
],remainder='passthrough')
```

```python
transformed.fit_transform(x_train).shape
```

```python
transformed.fit_transform(x_test).shape
```
