```python
# 1. Reverse a string
a='hello'
b=''

for i in a:
  b = i+b
print(b)
```

```python
# 2. Check palindrome

a='madam'
b=''

for i in a:
  b=i+b

if(a==b):
  print("true")
else:
  print("false")

```

```python
# 3. Count vowels in a string

a='anshu'
count=0

for i in a:
  if i in "aeiouAeiou":
    count+=1
print(count)
```

```python
# 4. Count consonants

a='python'
count=0

for i in a:
  if i not in "aeiouAeiou" and i != " ":
    count+=1
print(count)
```

```python
# 5. Convert string to uppercase

a='python'

print(a.upper())
```

```python
# 6. Convert string to lowercase

a='PYTHON'
print(a.lower())
```

```python
# 7. Find length of string (without len)

a='python'
count=0

for i in a:
  count+=1

print(count)
```

```python
# 8. Count spaces

a="hi how are you"
count=0

for i in a:
  if i == " ":
    count+=1
print(count)
```

```python
# 9. Remove spaces from string

a='h e l l o'
b=''

for i in a:
  if i != " ":
    b=b+i
print(b)

```

```python
# 10. Count frequency of a character

a='banana'
b= input("enter the character:")
count=0

for i in a:
  if i == b:
    count+=1
print(count)
```

```python
# 14. Replace spaces with hyphens

str='hello world'

for i in str:
  if i == " ":
    i = "-"
    print(i,end="")
  else:
    print(i,end="")
```

```python
# 15. Check if two strings are anagram

s1 = "listen"
s2 = "silent"

if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```

```python
# 16. Count words in a sentence

a='I love python'
count = 1

for i in a:
  if i == " ":
    count+=1
print(count)
```

```python
# 17. Find largest word in string

a='I love python'
b=a.split()

longest = ""

for i in b:
  if len(i)>len(longest):
    longest = i

print(longest)
```

```python
# 18. Remove all special characters

a='he@llo'

b='abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890'

for i in a:
  if i in b:
    print(i,end="")
```

```python
# 19. Check if string contains only digit

s = "1234"

if s.isdigit():
    print("Only digits")
else:
    print("Not digits")
```

```python
# 20. Capitalize first letter of each word

a = 'hello world'
s = a.split()

result = []

for i in s:
    new_word = i[0].upper() + i[1:]
    result.append(new_word)

print(result)
```
