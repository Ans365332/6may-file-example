```python
# Remove duplicate values from dictionary
d = {'a':1, 'b':2, 'c':1}
new={}

for k,v in d.items():
  if v not in new.values():
    new[k]=v
print(new)
```

```python
# Count number of words in a sentence using dictionary
sentence = input("Enter a sentence: ")
words = sentence.split()
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print(word_count)
```

```python
# Swap keys and values of dictionary
d={'a':1, 'b':2}
new={}
for k,v in d.items():
  new[v]=k
print(new)
```

```python
# Find common keys between two dictionaries
d1={'a':1,'b':2}
d2={'b':3,'c':4}
for k in d1:
  if k in d2:
    print(k)
```

```python
# Find second highest value key in dictionary
d={'a':10, 'b':50, 'c':30}

sort_d=sorted(d.items(), key=lambda x:x[1])
print(sort_d[-2])
```

```python
# Remove keys with None values
d = {'a':1, 'b':None, 'c':3}
new={}
for k,v in d.items():
  if v is not None:
    new[k]=v
print(new)
```

# Dictionary Comprehension

```python
# Create a dictionary where numbers from 1 to 5 are keys and their squares are values.

d={i:i**2 for i in range(1,6)}
d
```

```python
# Create a dictionary where numbers from 1 to 10 are keys and their cubes are values.

d={i:i**3 for i in range(1,11)}
d
```

```python
# Create a dictionary from a list of fruits where fruit name is key and its length is value.
fruits = ['apple', 'banana', 'cherry', 'date']
d={i:len(i) for i in fruits}
d
```

```python
# Create a dictionary where keys are numbers from 1 to 5 and values are "even" or "odd".
d={i:"even" if i%2==0 else "odd" for i in range(1,6)}
d
```

```python
# Convert this list into a dictionary where key is the character and value is ASCII code:
li=['a', 'b', 'c', 'd']
d={i:ord(i) for i in li}
d
```

```python
# Create a dictionary from a string where each character maps to its uppercase version.
# Example: "abc" → {'a':'A', 'b':'B', 'c':'C'}

s='abc'
d={i:i.upper() for i in s}
d
```

```python
# Swap keys and values of this dictionary using dictionary comprehension:

a={'a':1, 'b':2, 'c':3}

d={v:k for k,v in a.items()}
d
```

```python
# Create a dictionary where keys are numbers from 1 to 10 and values are only even numbers squared.

d={i:i**2 for i in range(1,11) if i%2==0}
d
```

```python
# Create a dictionary from two lists:
keys = ['name', 'age', 'city']
values = ['Anshu', 21, 'Jaipur']

d={keys[i]:values[i] for i in range(len(keys))}
d
```

```python
# Create a dictionary of numbers from 1 to 20 where only numbers divisible by 3 are included and value is square of the number.

d={i:i**2 for i in range(1,21) if i%3==0}
d
```

```python
# From this dictionary, create a new dictionary containing only items with value greater than 50:
a = {'a':10, 'b':60, 'c':30, 'd':90}
d={k:v for k,v in a.items() if v>50}
d
```

```python
# Create a dictionary where keys are words from a sentence and values are word lengths.
s="python is easy to learn"
a=s.split()
d={i:len(i) for i in a}
d
```

```python
# Create a dictionary where keys are numbers from 1 to 10 and values are:
# "Fizz" if divisible by 3
# "Buzz" if divisible by 5
# "FizzBuzz" if divisible by both
# otherwise the number itself

d={i:"FizzBuzz" if i%3==0 and i%5==0 else "Fizz" if i%3==0 else "Buzz" if i%5==0 else i for i in range(1,11)}
d
```

```python
# Convert this dictionary into uppercase keys:
a = {'name':'Anshu', 'city':'Delhi'}

d={k.upper():v for k,v in a.items()}
d
```

```python
# Create a dictionary where key is a number from 1 to 5 and value is a list of multiples of that number till 5 terms.

d={i:[i*j for j in range(1,6)] for i in range(1,6)}
d
```

```python
# Create a frequency dictionary for characters in a string using dictionary comprehension.
e="banana"
d={i:e.count(i) for i in e}
d
```

```python
# Create a nested dictionary where:
# keys are numbers from 1 to 3
# values are another dictionary containing square and cube.

d={i:{"square":i**2 , "cube":i**3} for i in range(1,4)}
d
```

```python
# From the given dictionary, remove all items whose values are strings:
s = {'a':10, 'b':'hello', 'c':25, 'd':'python'}

d={k:v for k,v in s.items() if type(v)==int}
d
```

```python
# Create a dictionary from a list of names where:
# key = name
# value = number of vowels in the name

names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']

d={i:sum(1 for j in i if j in 'aeiouAEIOU') for i in names}
d
```
