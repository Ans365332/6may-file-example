#Step-2--> Remove Html tags

```python
import re #regular expression
def remove_html_tags(text):
  pattern=re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
text= "<html><body><p> Movie1 </p><p> Actor - Aamir Khan </p><p> Click here to <a href='http://google.com'>download</a></p></body></html>"
```

```python
remove_html_tags(text)
```

```python
# now we remove tags from IMBD datasets
df['review']=df['review'].apply(remove_html_tags)
```

#Step-3 --> Remove Url

```python
# If i want to remove urls from text then

def remove_url(text):
  pattern=re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

```python
text2='For notebook click https://www.deture.com/delta/note1234acc'
```

```python
remove_url(text2)
```

```python
df['review']=df['review'].apply(remove_url)
```

#Step-4 --> Remove Punctuation

```python
import string
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
  for char in exclude:
    text=text.replace(char,'')
  return text
```

```python
text3= 'string. with. Punctuation?'
```

```python
remove_punc(text3)
```

```python
df['review']=df['review'].apply(remove_punc)
```

# Step-5--> Chatword Treatment

```python
chat_words ={'u2': 'you too', 'BBL':'Be Back Later','GAL':'Get A Life'}
```

```python
chat_words
```

```python
def chat_conversation(text):
  new_text = []
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append(chat_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

```python
chat_conversation('BBL')
```

# Step-7 Removing Stepwords

```python
import nltk # natural language tool kit
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
STOPWORDS = set(stopwords.words('english'))
```

```python
#STOPWORDS
```

```python
# manually work
def remove_stopwords(text):
  new_text =[]
  for word in text.split():
    if word in set(stopwords.words('english')):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return " ".join(x)

```

```python
remove_stopwords('probably my all-time favorite movie , a story of selflessness,sacrifice and dedication')
```

# Stemming

```python
# 1. Inflection ==> "In grammer, inflection is the modification of a word
# to express differrent grammatical categories such as tense , case , voice,
# aspect ,person , number , gender and mood"


# 2.Stemming ==> Stemming is the process of reducing inflecton in words to
# their root forms such as mapping a group of words to the same stem even if
# the stem itself is not a valid word in the language.
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
# stemming mapping a group of words to the same stem even if the stem itself
# is not valid word in language . so , this is the big issue of stemming


text = 'Probably my alltime favourite movie a story of selflessness sacrifise and dedication'
print(text)
```

```python
stem_words(text)
```

```python
# so , we have to use lemmatization instead of stemming
# note - stemming is fast and lemmatization is slow
```

```python
# Lemmatization , unlike stemming , reduces the inflected words properly
# ensuring that
# the root word belongs to the language.
# In lemmatization root word is called Lemma .
# Alemma(plural Lemmas or Lemmata) is the canonical form , dictionary form,
# or citation form of a set of words .config/


# stemming works on behalf of algorithm , so it is fast and Lemmatization is
# using wordnet package , so it is slow.
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python

wordnet_lemmatizer = WordNetLemmatizer()
#
```

```python
from nltk.tokenize.sonority_sequencing import punctuation
sentence = "He was running and eating at the same time . He has a bad habbit of swimming afte playing long hours in the sun"

punctuation='?:!.,;'
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
  sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
  print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```
