# Encoding --->nTransfrom your categorical data into numerical data
1. Label --> It will encode data one by one columns

2. Ordinal --> It will encode data more than one column

3. OneHot --> It will create new column based on subcategories but it has only values not columns name

4. get_dummies ---> It will create new column based on subcategories but it show column name with values

```python
# (1). LableEncoder
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/toyota - toyota.csv')
df.head()
```

```python
x= df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
x['model']=lb.fit_transform(x['model'])
x['transmission']=lb.fit_transform(x['transmission'])
x['fuelType']=lb.fit_transform(x['fuelType'])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```

# 2. OrdinalEncoder

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head()
```

```python
df=df.dropna()
```

```python
df=df.drop(columns=['age','fever'])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Male','Female'],['Mild','Strong'],
                              ['Kolkata','Delhi','Bangalore','Mumbai'],
                              ['Yes','No']])
```

```python
df_ar=oe.fit_transform(df)
```

```python
df_new=pd.DataFrame(df_ar,columns=df.columns)
```
