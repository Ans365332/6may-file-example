```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/insurance - insurance (1).csv')
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
```

```python
df.head(2)
```

```python
x=df.drop('charges',axis=1)
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# 4.Cars

```python
df=pd.read_csv('/content/cars - cars.csv')
df.head(2)
```

```python
df.info()
```

```python
df=df.drop(columns=['name'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df['fuel']=lb.fit_transform(df['fuel'])
df['seller_type']=lb.fit_transform(df['seller_type'])
df['transmission']=lb.fit_transform(df['transmission'])
df['owner']=lb.fit_transform(df['owner'])
```

```python
df.head(2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```
