You have to do these taks on this dataset :
1. import the data
2. Handle missing values if there are
3. Data Dividation using train_test_split
4. Convert Your x_train data into Noormal distribution using Standardization
5. Convert your normal distributed array into dataframe and then export this dataframe into excel  and csv format .

```python
import pandas as pd
import numpy as np
```

```python
# 1. import the data
df=pd.read_csv("/content/Churn_Modelling - Churn_Modelling.csv")
df.head()
```

```python
# 2. Handle missing values if there are

df.isnull().sum()
```

```python
df["Gender"] = df["Gender"].map({'Male':1,'Female':0})
df['Geography']=df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df.drop(columns=["Surname"],inplace=True)
```

```python
# 3. Data Dividation using train_test_split

x=df.drop(columns=["Exited"])
y=df["Exited"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
np.round(x_train.describe(),2)
```

```python
#4. Convert Your x_train data into Noormal distribution using Standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
```

```python
# 5. Convert your normal distributed array into dataframe and then export this dataframe into excel  and csv format .
x_train_new = pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```

```python
x_train_new.to_csv("x_train_new.csv", index=False)
```
