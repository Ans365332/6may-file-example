```python
# Find length of a list without using len()

a=[1,2,3,4]
count=0

for i in a:
   count+=1
print(count)

```

```python
# Find sum of all elements without using sum()

a=[1,2,3,4]
sum=0

for i in a:
  sum+=i
print(sum)
```

```python
# Find average of list elements

a=[1,2,3,4]
sum=0
count=0

for i in a:
  sum+=i
  count+=1
print(sum/count)
```

```python
# Remove last element from list manually

a=[1,2,3,4]
a=a[:-1]
print(a)
```

```python
# Count occurrences of a number in a list

a=[1,2,1,2,3,4,5,3,4,5,2]
count=0

b=2
for i in a:
  if i==b:
    count+=1
print(count)
```

```python
# Find index of a given element manually

a=[1,2,3,4,5,6,7]
b=int(input("Enter the number: "))
for i in range(len(a)):
  if a[i]==b:
    print(i)
```

```python
# Check if an element exists in list

a=[1,2,3,4,5,6,7]
b=int(input("Enter the number: "))
if b in a:
  print("Element exists")
else:
  print("Element does not exist")
```

```python
# Sort a list without using sort()

a=[3,5,2,6,4,1]
for i in range(len(a)):
  for j in range(i+1,len(a)):
    if a[i]>a[j]:
      a[i],a[j]=a[j],a[i]
print(a)
```

```python
# Reverse a list without using reverse() or slicing

a=[1,2,3,4,5]
b=[]

for i in a:
  b.append(a[-i])
print(b)
```

```python
# Print all even numbers from a list

a=[1,2,3,4,5,6,7,8]
b=[]
for i in a:
  if i%2==0:
    b.append(i)
print(b)
```

```python
# Print all odd numbers from a list
a=[1,2,3,4,5,6,7,8]
b=[]
for i in a:
  if i%2!=0:
    b.append(i)
print(b)
```

```python
# Find maximum element without using max()

a=[1,4,5,3,6,7,2]

a.sort()
print(a[-1])
```

```python
# Find minimum element without using min()

a=[1,4,5,3,6,7,2]

a.sort()
print(a[0])
```

```python
# Find the square of each element in a list

a=[1,2,3,4]
b=[]
for i in a:
  b.append(i**2)
print(b)
```

```python
# Multiply all elements of a list

a=[1,2,3,4]
b=[]
for i in a:
  b.append(i*i)
print(b)
```

```python
# Replace all negative numbers with 0

a=[1,-2,3,-4]
for i in range(len(a)):
  if a[i]<0:
    a[i]=0
print(a)
```

```python
# Separate positive and negative numbers into two lists

a=[1,-2,3,-4]
p=[]
n=[]
for i in a:
  if i>0:
    p.append(i)
  else:
    n.append(i)
print(p)
print(n)
```

```python
# Find the first repeating element in a list

a=[1,2,3,3,4,5,5,6]

for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i]==a[j]:
            print(a[i])
            break
        else:
            continue
```
