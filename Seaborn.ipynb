# Seaborn-->

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Getting the dataset which are into the seaborn

sns.get_dataset_names()
```

```python
# Loading the dataset
df=sns.load_dataset('tips')
df
```

```python
# Scatter plot in seaborn on total bill and tip

plt.figure(figsize=(5,3)) # -- it will give the size
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title("Plotting the scatter plot between the total bill and tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Line plot used for the trends and continuous data(size,tip)
plt.figure(figsize=(5,3))
sns.lineplot(x='size',y='tip',data=df)
plt.title("Plotting the line plot between the size and tip")
plt.xlabel("Size")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.lineplot(x='size',y='tip',data=df,marker='o')
plt.title("Plotting the line plot between the size and tip")
plt.xlabel("Size")
plt.ylabel("Tip")
plt.show()
```

```python
# Barplot used to show the value with confidence interval(total bill,day)
# time will always come on x axis
plt.figure(figsize=(5,3))
sns.barplot(x='day',y='total_bill',data=df)
plt.title("Plotting the bar plot between the day and total bill")
plt.xlabel("day")
plt.ylabel("Total Bill")
plt.show()
```

-- here in bar plot | on bar are error bar that show uncertanity(inc and dec) in data

```python
df.info()
```

```python
# countplot--> count frequency of categories(like histogram but for category )

plt.figure(figsize=(5,3))
sns.countplot(x='day',data=df)
plt.title("Plotting the count plot between the day")
plt.xlabel("day")
plt.ylabel("Frequency")
plt.show()
```

# Insights of day countplot-->
1.on day thrusday frequency is approx 65

2.on day friday frequency is approx 18 that mean on friday we get low amount of tips

3.on saturday frequency is ver high that is 80 above that mean on saturday we get high amount of tips

4.on sunday freuency is slightly lower than Saturday and greater than Thrusday aprrox 70 that mean on sunday we get tips lower than saturday

```python
plt.figure(figsize=(5,3))
sns.countplot(x='smoker',data=df)
plt.title("Plotting the count plot between the smoker")
plt.xlabel("smoke")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='time',data=df)
plt.title("Plotting the count plot between the time")
plt.xlabel("time")
plt.ylabel("Frequency")
plt.show()
```

```python
# Distribution + Outlier (boxplot)

plt.figure(figsize=(5,3))
sns.boxplot(x='day',y='total_bill',data=df)
plt.title("Plotting the box plot between the day and total bill")
plt.xlabel("day")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x='total_bill',data=df)
plt.title("Plotting the box plot of  total bill")
plt.xlabel("Total bill")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x='size',data=df)
plt.title("Plotting the box plot  size")
plt.xlabel("size")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(5,3))
arg,axs=plt.subplots(2,2)
axs[0,0].boxplot(df['total_bill'])
axs[0,1].boxplot(df['size'])
axs[1,0].scatter(df['total_bill'],df['tip'])
axs[1,1].scatter(df['size'],df['tip'])
plt.show()
```
