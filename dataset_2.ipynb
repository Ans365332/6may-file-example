```python
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic - titanic.csv')
df
```

# Data Cleaning & Basic Exploration

```python
# 1.Check the first 5 rows.
df.head()
```

```python
# 2. Find the shape (rows & columns) of the dataset.
df.shape
```

```python
# 3. Get column names and their data types.
df.dtypes
```

```python
# 4. Check for missing values in each column.
df.isna().sum()
```

```python
# 5. Drop the column Ticket (since it is not useful for analysis).
df.drop('Ticket',axis=1,inplace=True)
```

```python
df
```

```python
# 6. Fill missing values in Age with the mean age.
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Age'].isna().sum()
```

```python
# 7. Fill missing values in Embarked with the most frequent value.
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df['Embarked'].isna().sum()
```

```python
# 8. Create a new column FamilySize = SibSp + Parch + 1.
df['FamilySize']=df['SibSp']+df['Parch']+1
```

```python
df
```

```python
# 9. Drop the column Cabin (too many missing values).
df.drop('Cabin',axis=1,inplace=True)
```

```python
df
```

```python
# 10. Rename the column Pclass to PassengerClass.
df.rename(columns={'Pclass':'PassangerClass'},inplace=True)
```

```python
df
```

# Filtering & Selection

```python
# 1. Select all passengers who were in PassengerClass 1.
df[df['PassangerClass']==1]
```

```python
# 2. Select all passengers who are older than 50 years.
df[df['Age']>50]
```

```python
# 3. Find all female passengers who survived.
df[(df['Sex']=='female')&(df['Survived']==1)]
```

```python
# 4. Find all passengers who paid a fare greater than 50.
df[df['Fare']>50]
```

```python
# 5. Find the passengers who were traveling alone (FamilySize == 1).
df[df['FamilySize']==1]
```

# Grouping & Aggregation

```python
# 1. Find the average age of passengers grouped by Sex.
avg=df.groupby('Sex')['Age'].mean()
avg
```

```python
# 2. Find the survival rate grouped by PassengerClass.
avg=df.groupby('PassangerClass')['Survived'].mean()
avg
```

```python
# 3. Find the maximum fare paid by each PassengerClass.
max=df.groupby('PassangerClass')['Fare'].max()
max
```

```python
# 4. Count how many passengers boarded from each port (Embarked).
df['Embarked'].value_counts()
```

```python
avg=df.groupby('Survived')['FamilySize'].mean()
avg
```
