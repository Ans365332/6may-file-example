```python
import numpy as np
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/Social_Network_Ads - Social_Network_Ads.csv')
df.head()
```

```python
df=df.drop(columns=['User ID','Gender'])
```

```python
x=df.drop(columns=['Purchased'],axis=1)
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=42)
```

```python
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('pca',PCA(n_components=2)),
    ('classifier',RandomForestClassifier(n_estimators=100,random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

# for user behavior data

```python
df=pd.read_csv('/content/userbehaviour - userbehaviour.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['Status']=lb.fit_transform(df['Status'])
```

```python
x=df.drop(columns=['userid','Status'],axis=1)
y=df['Status']
```

```python
# now we will use smote
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE()
```

```python
x_res,y_res=smote.fit_resample(x,y)
```

```python
x_res_train,x_res_test,y_res_train,y_res_test=train_test_split(x_res,y_res,test_size=0.2,random_state=42)
```

```python
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('pca',PCA(n_components=2)),
    ('classifier',RandomForestClassifier(n_estimators=100,random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_res_train,y_res_train)
```

```python
y_pred=pipe.predict(x_res_test)
```

```python
acc=accuracy_score(y_res_test,y_pred)
acc
```
