# Column Transformer-->

```python
# without column transformer
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('/content/covid_toy - covid_toy.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
x_train
```

# Manually type output

```python
# adding simple imputer to fever column

si=SimpleImputer(strategy = "mean")
x_train_fever=si.fit_transform(x_train[['fever']])

# also the test data
x_test_fever=si.fit_transform(x_test[['fever']])


x_train_fever.shape
```

```python
# Ordinal Encoding ---> Cough

oe = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough=oe.fit_transform(x_train[['cough']])


# also the test data
x_test_cough=oe.fit_transform(x_test[['cough']])

x_train_cough.shape
```

```python
# OneHot Encoding ---> Gender,City

ohe=OneHotEncoder(drop='first',sparse_output= False)

x_train_gender_city=ohe.fit_transform(x_train[['gender','city']])


# also the test data
x_test_gender_city=ohe.fit_transform(x_test[['gender','city']])
x_train_gender_city.shape
```

```python
# Extracting Age

x_train_age=x_train.drop(columns=['gender','fever','cough','city']).values

# also the test data

x_test_age=x_test.drop(columns=['gender','fever','cough','city']).values

```

```python
x_train_age.shape
```

```python
x_train_transformed=np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
```

```python
x_test_transformed=np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis=1)
```

```python
x_train_transformed.shape
```

# By the help of column Transformer

```python
from sklearn.compose import ColumnTransformer  # this is how to import columntransformer

transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),  # in a 'fever' column by the
    # the help of SI we fill missing values by mean , meadian , mode.
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    # by this process we encode our data.
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')

# remainder = passthrough ==> it means rest all the columns remain same.

```

```python
transformer.fit_transform(x_train).shape
```

```python
transformer.fit_transform(x_test).shape
```

# with column transformer

```python
from sklearn.compose import ColumnTransformer  # this is how to import columntransformer

transformer=ColumnTransformer(transformers=[
    ('tnf1',OrdinalEncoder(categories=[['No','Yes']]),['smoker']),

    ('tnf2',OrdinalEncoder(categories=[['Male','Female']]),['sex']),
    # by this process we encode our data.

    ('tnf3',OrdinalEncoder(categories=[['Dinner','Lunch']]),['time']),
    ('tnf4',OneHotEncoder(sparse_output=False,drop='first'),['day'])
],remainder='passthrough')
```
