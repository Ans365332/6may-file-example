```python
from googletrans import Translator
trans = Translator()

text = 'mujhe python pasand hai'

translated = trans.translate(text, src = 'hi',dest='en')
print(translated.text)
```

```python

from googletrans import Translator


translator = Translator()


text = """
Aaj kal technology humari life ka bahut important part ban gayi hai.
Har field me AI aur data science ka use ho raha hai.
Students bhi ab online platforms se coding seekh rahe hain.
"""


translated = translator.translate(text, dest='en')


print(translated.text)
```

```python
new=translated.text.split('.')
```

```python
new
```

```python
english_list = [i.strip() for i in new if i.strip()]
```
