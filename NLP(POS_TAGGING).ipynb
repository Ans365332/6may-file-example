```python
# POS Tagging in NLP

# POS Tagging means part of speech tagging.
# It is an NLP Technique where each word in sentence is assigned its grammatical role such as:
# Noun
# Verb
# Adjective
# Adverb
# Conjunction
# Preposition
# Pronoun
```

```python
# Why POS Tagging is Important

# POS Tagging is used in:
# Text Classification
# Chatbots
# Machine Translation
# Named Entity recognition(NER)
# Sentiment Analysis
# Speech Recognition
# Grammar Checking
```

```python
# Common POS tags
```

```python
POS_tags = Meaning

NN = Noun, singular or mass
NNS = Noun, plural Noun
NNP = Proper noun, singular
NNPS = Proper noun, plural
VB = Verb
VBD = Verb, past tense
VBG = Verb, gerund
JJ = Adjective
RB = Adverb
PRP = Pronoun
IN = Preposition
```

```python
# Working process of POS Tagging ::::

# 1. Sentence is divided into tokens.
# 2. NLP model checks grammar/context.
# 3. Appropriate POS tag is assigned to each word.
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/IMDB Dataset - IMDB Dataset.csv")
df.head()
```

```python
# now i have to clean this data

# 1. lowercase

df['review'] = df['review'].str.lower()
```

```python
df.head()
```

```python
# 2. Remove html tags
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'', text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
# 3. Remove url

import re

def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'', text)
```

```python
df['review'] = df['review'].apply(remove_url)
```

```python
df.head()
```

```python
# 3. remove punctuation

import string
string.punctuation

exclude = string.punctuation

def remove_punc(text):
  for char in exclude:
    text = text.replace(char, '')
  return text
```

```python
df['review'] = df['review'].apply(remove_punc)
```

```python
# remove stopword
import nltk # natural language tool kit
nltk.download('stopwords')
from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))

# manually work
def remove_stopwords(text):
  new_text =[]
  for word in text.split():
    if word in set(stopwords.words('english')):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return " ".join(x)

```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
x = df['review'][3]
x
```

```python
# NER


import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp(x)

for token in doc.ents:
  print(token.text, "--->", token.label_)
```

```python
from spacy import displacy
displacy.serve(doc, style="ent")
```
