```python
import pandas as pd
```

```python
df=pd.read_csv('/content/kidney_disease - kidney_disease.csv')
df
```

```python
# 1. Group patients by classification and calculate the average age.

avg=df.groupby('classification')['age'].mean()
avg
```

```python
# 2. Group by classification and count how many patients are in each class.

count=df.groupby('classification')['id'].count()
count
```

```python
# 3. Group by htn (hypertension) and calculate the mean bp.

mean=df.groupby('htn')['bp'].mean()
mean
```

```python
# 4. Group by dm (diabetes) and count how many patients fall into each category.
count=df.groupby('dm')['id'].count()
count
```

```python
# 5. Group by appet (appetite) and find the average hemo.
avg=df.groupby('appet')['hemo'].mean()
avg
```

```python
# 6. Group by rbc and calculate the mean of pcv.

df['pcv']=pd.to_numeric(df['pcv'],errors='coerce')
df['pcv']
mean=df.groupby('rbc')['pcv'].mean()
mean

```

```python
# 7. Group by pc (pus cell) and count how many patients are in each group.

count=df.groupby('pc')['id'].count()
count
```

```python
# 8. Group by classification and calculate the average sc (serum creatinine).

avg=df.groupby('classification')['sc'].mean()
avg
```

```python
# 9. Group by classification and find the maximum age in each group.
max=df.groupby('classification')['age'].max()
max
```

```python
# 10. Group by both htn and dm and count patients in each subgroup.
count=df.groupby(['htn','dm'])['id'].count()
count
```

# Dropna

```python
# 1. Drop rows where age is missing.
df1=df.dropna(subset=['age'])
df1
```

```python
# 2. Drop rows where either bp or sg is missing.
df2=df.dropna(subset=['bp','sg'])
df2
```

```python
# 3. Drop rows where all values are missing.
df3=df.dropna()
df3
```

```python
# 4. Drop rows where rc (red blood cell count) is missing.
df4=df.dropna(subset=['rc'])
df4
```

```python
# 5. Create a new DataFrame without missing values in hemo.
df5=df.dropna(subset=['hemo'])
df5
```

# Sum-

```python
# 1. Find the sum of bp for all patients.
df['bp'].sum()
```

```python
# 2. Calculate the sum of age grouped by classification.
df.groupby('classification')['age'].sum()
```

```python
# 3. Find the total number of missing values in the dataset (df.isna().sum().sum()).
df.isna().sum().sum()
```

```python
# 4. Sum up pcv values grouped by htn.
df.groupby('htn')['pcv'].sum()
```

```python
# 5. Find the sum of wc (white blood cell count) for diabetic patients.
df['wc']=pd.to_numeric(df['wc'],errors='coerce')
df.groupby('dm')['wc'].sum()
```
