```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/cars - cars.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['fuel'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['name']=le.fit_transform(df['name'])
df['fuel']=le.fit_transform(df['fuel'])
df['transmission']=le.fit_transform(df['transmission'])
df['owner']=le.fit_transform(df['owner'])
df['seller_type']=le.fit_transform(df['seller_type'])
```

```python
x=df.drop(columns=['selling_price'])
y=df['selling_price']
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
x_train_new.describe()
```

```python
df=pd.read_csv('/content/credit_scoring - credit_scoring (1).csv')
df.head(2)
```

```python
df['Type of Loan'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Male','Female'],['Married','Divorced','Single'],
                              ['Master','High School','PhD','Bachelor'],
                              ['Self-Employed','Employed','Unemployed'],
                              ['Auto Loan','Home Loan','Personal Loan']])
```

```python
df_ar=oe.fit_transform(df[['Gender','Marital Status','Education Level','Employment Status','Type of Loan']])
```

```python
df_new=pd.DataFrame(df_ar,columns=['Gender','Marital Status','Education Level','Employment Status','Type of Loan'])
```
