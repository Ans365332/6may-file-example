```python
# Write a decorator that prints "Function Started" before executing a function

def start_decorator(func):
    def wrapper():
        print("Function Started")
        func()
    return wrapper

@start_decorator
def greet():
    print("Hello")

greet()
```

```python
# Write a decorator that prints "Function Ended" after a function finishes execution.

def start_deco(func):
  def wrapper():
    func()
    print("Function Ended")
  return wrapper

@start_deco
def greet():
  print("Hello")

greet()
```

```python
# Write a decorator that prints both "Before Function" and "After Function" around a function call.

def star_deco(func):
  def wrapper():
    print("Before function")
    func()
    print("after function")
  return wrapper

@star_deco
def greet():
  print("Hello")

greet()
```

```python
# Write a decorator that works with a function that takes one argument and prints the argument before execution.

def show_argument(func):
  def wrapper(name):
    print(name)
    func(name)
  return wrapper

@show_argument
def greet(name):
  print(f"Hello {name}")

greet("John")
```
